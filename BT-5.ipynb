W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def kiemtra(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
k=int(input())
a= list(map(int,input().split()))
def tim_vitri_dau(a,k):
    for i in range(len(a)):
        if a[i]==k:
            return i
    return -1
print(tim_vitri_dau(a, k))
    
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 


```python
n=int(input())
def giaithua(n):
    if n==0 or n==1:
        return 1
    return n * giaithua(n-1)  #Hàm đệ quy
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.


```python
def divide_with_remainder(dividend, divisor):
    if divisor ==0:
        return ("không chia cho 0")
    q = dividend // divisor
    r = dividend % divisor
    return (q,r)
    
      
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op.


```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return "Không chia được cho 0"
        return a / b
    else:
        return "Toán tử không hợp lệ"
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def tam_giac_vuong(a,b,c):
    max_canh=max(a,b,c)
    if max_canh == a:
        return b**2 + c**2 == a**2
    elif max_canh == b:
        return a**2 + c**2 == b**2
    else:
        return b**2 + a**2 == c**2
    
```
