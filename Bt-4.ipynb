W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python

def collatz_length(x):
    count = 1
    while x != 1:
        if x%2 ==0:
            x//2 ==2
        else:
            x = 3*x +1
        count +=1
    return count
n = int(input())
max_len=0 # độ dài lớn nhất tìm được
best_x =1 # số tương ứng
for i in range (1,n+1):
    length= collatz_length(i)
    if length > max_len:
        max_len = length
        best_x = i
    elif length == max_len and i < best_x:
        best_x= i
print(best_x, max_len)


```
